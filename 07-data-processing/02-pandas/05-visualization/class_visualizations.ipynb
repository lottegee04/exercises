```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
#sample data

data  = {
    'year' : ['2020', '2021', '2022' , '2023', '2024'],
    'Pigeons': [33,52,26,40,38],        
    'Sparrows': [54,48,66,59,61],
    'others' : [104,98,88,92,101]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
plt.figure()
plt.plot(df['year'],df['Pigeons'], marker = "o", color = "blue", markersize = 8, label = "Pigeons", linewidth = 2)
plt.plot(df['year'], df['Sparrows'], linestyle = 'dotted', color = 'green', marker = '>', markeredgecolor = 'red', label ="Sparrows" )
plt.plot(df['year'], df["others"], linestyle = "dashed", color =  "pink", marker ="*", gapcolor = "purple", label = 'others')

df['total'] = df['Pigeons'] + df['Sparrows'] + df['others']

plt.bar(df["year"], df["total"], color = 'purple' , label = 'total', width = 0.4 )
plt.title("Aantal vogels doorheen de jaren")
plt.xlabel("Jaartal")
plt.ylabel("Aantal")
plt.legend()
```

```python
plt.figure()
plt.bar(df['year'],df['Pigeons'], color = "pink")
plt.bar(df['year'], df['Sparrows'], color = "purple", bottom= df['Pigeons'])
plt.bar(df['year'], df["others"], color =  "blue", bottom= df["Sparrows"]+df['Pigeons'])
plt.xlabel("Years")
plt.ylabel("Count")
```
